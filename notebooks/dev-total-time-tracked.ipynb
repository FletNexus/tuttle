```python
def total_time_tracked(
    by: str
) -> pandas.DataFrame:
    """Calculate the total time spent, grouped by project, client..."""
    if by == "project":
        raise NotImplementedError()
    elif by == "client":
        raise NotImplementedError()
    else:
        raise ValueError()
```

```python
from tuttle.cloud import login_iCloud
```

```python
icloud = login_iCloud(
    ICloudAccount(
        user_name=input("user name:")
    )
)
```

```python
from tuttle.calendar import CloudCalendar
```

```python
tracking_cal = CloudCalendar(
    name="TimeTracking",
    icloud=icloud
)
```

```python
tracking_cal.to_data()
```

```python
time_by_project = (
    tracking_cal
    .to_data()
    .filter(["title", "duration"])
    .groupby("title")
    .sum()
)
hours_by_project = (time_by_project["duration"] / pandas.Timedelta('1 h')).to_frame()
hours_by_project
```

```python
import altair
```

```python
from tuttle.dataviz import enable_dark_theme
enable_dark_theme()
```

```python
(
    altair.Chart(
        (
            hours_by_project
            .reset_index()
        )
    )
    .mark_bar()
    .encode(
        x='title',
        y='duration'
    )
    .properties(
        width=600,
        height=200,
        title="hours by project"
    )
    .interactive()
)
```
